# Milestone 5 Harmonization And Overlap

Aggregate-only review notebook for `reports/harmonization_manifest.json`, `reports/harmonization_coverage.json`, and `reports/unmapped_concepts.json`. Do not load patient-level harmonized Parquet rows in this notebook.

```python
import json
from pathlib import Path

REPORTS_ROOT = Path("..") / "reports"


def load_json(name):
    path = REPORTS_ROOT / name
    if not path.exists():
        return {"missing_report": str(path)}
    return json.loads(path.read_text())


manifest = load_json("harmonization_manifest.json")
coverage = load_json("harmonization_coverage.json")
unmapped = load_json("unmapped_concepts.json")
```

```python
artifact_rows = []
for name, artifact in manifest.get("artifacts", {}).items():
    artifact_rows.append(
        {
            "artifact": name,
            "rows": artifact.get("rows"),
            "path": artifact.get("path"),
        }
    )
artifact_rows
```

```python
coverage_by_domain = {}
for row in coverage.get("coverage", []):
    coverage_by_domain.setdefault(row.get("domain"), []).append(row)
coverage_by_domain
```

```python
unit_compatibility = coverage.get("unit_compatibility", [])
unit_compatibility
```

```python
unmapped_summary = {
    "data_safety": unmapped.get("data_safety"),
    "unmapped_groups": len(unmapped.get("unmapped", [])),
}
unmapped_summary
```
